```python
#how to return multiple values
def add1():
    n1=int(input("enter"))
    n2=int(input("enter"))
    
    sum=n1+n2
    mul=n1*n2
    sub=n1-n2
    div=n1/n2
    return sum,sub,mul,div
     
result=add1()
print(result)
```

```python
mylist=[1,4,5,2,7,9]
def serach(target,mylist):
    for i in mylist:
        if i==target:
            print("founded target",i)
            return i
    
    return -1
tar=int(input("enter"))
ans=serach(tar,mylist)
ans
```
