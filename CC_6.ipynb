```python
#given an array return an array where each element is the product of all the element in the array except itself
def product_(nums):
    n = len(nums)
    res = [1] * n
    
    left = 1
    for i in range(n):
        res[i] = left
        left *= nums[i]
    
    right = 1
    for i in range(n - 1, -1, -1):
        res[i] *= right
        right *= nums[i]
    
    return res
product_([1,2,3,4])
```
