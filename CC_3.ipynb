```python
name2="rudramandbravo"
count_const=0
count_vowel=0
for i in name2 :
    if(i=='a' or i=='e' or i=='i' or i=='o' or i=='u'):
        count_vowel=count_vowel+1
    else:
        count_const=count_const+1
print(count_vowel)
print(count_const)
```

```python
data=['a','e','i','o','u']
v_count=0
c_count=0
for i in name2:
    if i in data:
        v_count+=1
    else:
        c_count+=1
print(v_count)
print(c_count)
```

```python
name4 = []
for i in range(len(name2)):
    found = False
    for j in range(len(name4)):
        if name2[i] == name4[j]:
            found = True
    if not found:
        name4.append(name2[i])

print(name4)
```
